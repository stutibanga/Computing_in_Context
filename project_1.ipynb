# Calculating the Minimum dietary energy requirement (kcal/cap/day) for different countries in 2022

- Step 1: Display the dataset (should have atleast 1000 rows)

```python
import pandas as pd
```

```python
df = pd.read_csv("dietry_req.csv")
df
```

- Step 2: Filter the dataset to only compute the dietary requirements for all countries for the year 2022

```python
df_2022 = df[df['Year Code'] == 2022]
df_2022
```

- Step 3: Find the Mean, Median and Mode using pandas

```python
mean_value = df_2022['Value'].mean()
median_value = df_2022['Value'].median()
mode_value = df_2022['Value'].mode()[0]

print(f"Mean of the minimum dietary energy requirement is: {mean_value:.2f}(kcal/cap/day)")
print(f"Median of the minimum dietary energy requirement is: {median_value}(kcal/cap/day)")
print(f"Mode of the minimum dietary energy requirement is: {mode_value}(kcal/cap/day)")
```

- Step 4: Find the Mean, Median, Mode without pandas

```python
import csv
from statistics import mean, median, mode
```

```python
with open('dietry_req.csv', 'r') as file:
    reader = csv.DictReader(file)
    data_2022 = [float(row ['Value'])
                 for row in reader if row['Year'] == '2022']
```

```python
mean_value = mean(data_2022)
median_value = median(data_2022)
mode_value = mode(data_2022)

print(f"Mean of the minimum dietary energy requirement is: {mean_value:.2f}(kcal/cap/day)")
print(f"Median of the minimum dietary energy requirement is: {median_value}(kcal/cap/day)")
print(f"Mode of the minimum dietary energy requirement is: {mode_value}(kcal/cap/day)")
```

- Step 5: Data Visualization

```python
import csv
from statistics import mean, median, mode
```

```python
with open('dietry_req.csv', 'r') as file:
    reader = csv.DictReader(file)
    data_2022 = [
        (row['Area'], float(row['Value']))
        for row in reader if row['Year'] == '2022'
        ]
```

```python
values = [value for _, value in data_2022]
mean_value = mean(values)
median_value = median(values)
mode_value = mode(values)

print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value:.2f}")
```

```python
print("\nMinimum Dietary Energy Requirement (kcal/cap/day) for 2022")
max_value = max(values)
bar_width = 50 


for country, value in data_2022:
    bar_length = int((value / max_value) * bar_width)
    bar = '*' * bar_length
    print(f"{country}: {bar} {value:.2f}")
```
